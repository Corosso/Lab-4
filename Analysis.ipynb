```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbs
df = pd.read_excel("https://github.com/Corosso/Lab-4/raw/main/data.xlsx")
```

2)

```python
dfAmericasAvrgInc=df[(df["whoreg6"]=="Americas") & (df["wbincome2023"]=="Upper middle income") & (df["indicator_name"]=="Medical doctors (%)")]
dfAmericasAvrgInc = dfAmericasAvrgInc.sort_values(by='setting', key=lambda col: col != 'Colombia')
dfAmericasAvrgInc
```

```python
sbs.set(style="whitegrid")
plt.figure(figsize=(15, 6))
sbs.barplot(x='setting', y='estimate', hue='subgroup', data=dfAmericasAvrgInc, palette='muted')
plt.title('Gráfico de Barras por País y Sexo')
plt.xlabel('País')
plt.ylabel('Indicador en el eje Y')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
dfAvgSixRegions=df[(df["indicator_name"]=="Medical doctors (%)")]
dfColombia= dfAvgSixRegions[dfAvgSixRegions['setting'] == 'Colombia']
dfColombia=dfColombia.groupby(["setting",'whoreg6',"subgroup"])['estimate'].mean().reset_index()
dfAvgSixRegions=dfAvgSixRegions.groupby(['whoreg6',"subgroup"])['estimate'].mean().reset_index()
dfColombia['whoreg6'] = dfColombia['setting']
dfAvgSixRegions = pd.concat([dfColombia, dfAvgSixRegions])
dfAvgSixRegions
```

```python
sbs.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sbs.barplot(x='whoreg6', y='estimate', hue='subgroup', data=dfAvgSixRegions, errorbar=None)
plt.title('Gráfico de Barras por País y Sexo')
plt.xlabel('País/Region')
plt.ylabel('Indicador en el eje Y')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

¿Cómo se sitúa Colombia en el entorno regional e internacional frente a cada variable que  se  pregunta?
Análisis de brecha en formación médica por sexo:
Vemos que Colombia frente a algunos paises de America con ingresos similares, tiene una brecha similar, estos paises son, republica dominicana y paraguay.
Sin embargo el resto de paises tienen una brecha similar, es decir Colombia es un pais que tiene una brecha muy alta comparada con otros paises de la region
Ahora frente a las 6 regiones establecidas por la OMS, se ve reflejado que colombia es un pais con tendencias diferentes, ya que frente al promedio de las 6 regiones establecidas por la OMS, es un pais donde predomina la formacion medica para las mujeres, ya que en otras regiones, esta profesion la predominan los hombres, excetuando europa y america, donde esta brecha es muy similar
Análisis  de  gasto  en  salud  por  hogares:
Frente a los paises con ingresos similares, podemos ver que casi todos los paises tienen un porcenraje alto, colombia se encuentra cerca de la media de todos estos paises.
Se evidencia que colombia tiene un porcenatje relativamente alto frente a las regiones estabcidas por la OMS, sin embargo se puede llegar a comprar con la region del meditarraneo orientak, ya que sus porcentajes son casi iguales.


```python
df_gastos = df[(df["indicator_name"] == "Population pushed below a relative poverty line by household health expenditures (60% of median daily per capita consumption or income) (%)")]
df_colombia = df_gastos[df_gastos['setting'] == 'Colombia']
df_colombia = df_colombia.groupby(["setting", 'whoreg6', "subgroup"])['estimate'].mean().reset_index()
df_gastos = df_gastos.groupby(['whoreg6', "subgroup"])['estimate'].mean().reset_index()
df_colombia['whoreg6'] = df_colombia['setting']
df_gastos = pd.concat([df_colombia, df_gastos])
df_gastos
```

```python
sbs.set(style="whitegrid")
plt.figure(figsize=(20, 8))
sbs.barplot(x='whoreg6', y='estimate', hue='subgroup', data=df_gastos, palette='muted')
plt.title('Porcentaje de población debajo del umbral de pobreza por región e internacionalmente')
plt.xlabel('internacional')
plt.ylabel('Porcentaje de población debajo del umbral de pobreza')
plt.legend(bbox_to_anchor=(0.05, 1), loc='upper left')
plt.show()
```

```python
sbs.set(style="whitegrid")
plt.figure(figsize=(20, 8))
sbs.barplot(x='setting', y='estimate', hue='subgroup', data=df_gastos, palette='muted')
plt.title('Porcentaje de población debajo del umbral de pobreza por regional')
plt.xlabel('Regional')
plt.ylabel('Porcentaje de población debajo del umbral de pobreza')
plt.legend(bbox_to_anchor=(0.05, 1), loc='upper left')
plt.show()
```
